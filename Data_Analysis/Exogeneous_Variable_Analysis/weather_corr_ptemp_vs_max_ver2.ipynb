```python
import pandas as pd
import numpy as np
from scipy.stats import linregress
import matplotlib.pyplot as plt
import scipy.stats as stats

from sklearn import datasets
data = pd.read_excel("jeju_asos_data.xlsx")

idx = 0
corr_result = []  # 상관계수를 저장할 리스트
pval_result = []  # p-value를 저장할 리스트
#step = 37
stop = 0
cnt = 1

while stop == 0 :

  if cnt%12 == 1:
    step = 31
  elif cnt%12 == 2:
    step = 28
  elif cnt%12 == 3:
    step = 31
  elif cnt%12 == 4:
    step = 30
  elif cnt%12 == 5:
    step = 31
  elif cnt%12 == 6:
    step = 30
  elif cnt%12 == 7:
    step = 31
  elif cnt%12 == 8:
    step = 31
  elif cnt%12 == 9:
    step = 30
  elif cnt%12 == 10:
    step = 31
  elif cnt%12 == 11:
    step = 30
  else :
    step = 31

  if idx + step >= 1095 :
    df = data.iloc[idx:1095]
    stop = 1
  else :
    df = data.iloc[idx:idx+step]  # step개씩 잘라서 분석
    idx += step
    cnt += 1

  start_date =df['date'].values[0]
  end_date = df['date'].values[-1]
  max = df['max'].values
  #min = df['min'].values
  ptemp = df['p-temp'].values
  #temp = df['temp'].values
  #rain = df['rain'].values
  #humidity = df['humidity'].values
  #solar = df['solar'].values
  #cloud = df['cloud'].values
  #wind = df['wind'].values

  print('\n')
  print('<' + str(start_date) + ' ~ ' + str(end_date) + '>')
  x = ptemp
  y = max

  plt.scatter(x, y, alpha=0.5)
  plt.title('ptemp vs max')
  plt.xlabel('ptemp')
  plt.ylabel('max')
  slope, intercept, r_value, p_value, std_err = linregress(x, y)
  plt.plot(x, slope*x + intercept, color='red', label='Trendline')
  plt.show()

  print(stats.pearsonr(x,y))

  corr = stats.pearsonr(x,y)[0]
  pvalue = stats.pearsonr(x,y)[1]
  corr_result.append(corr) # 상관계수를 리스트에 추가
  pval_result.append(pvalue) # p-value를 리스트에 추가

print("\nCorrelation Results:")
print(corr_result)

print("\np-value Results:")
print(pval_result)
```
